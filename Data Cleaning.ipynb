```python
import pandas as pd
```

```python
# Try with 'latin-1' encoding (often works with Excel files)
df = pd.read_csv('lemur_data.csv', encoding='latin-1')

# Or try with 'iso-8859-1' encoding
df = pd.read_csv('lemur_data.csv', encoding='iso-8859-1')

# Or try with 'cp1252' encoding (Windows default)
df = pd.read_csv('lemur_data.csv', encoding='cp1252')
```

```python
df_cleaned = df.copy()
df_cleaned.head()
```

```python
df_cleaned = df_cleaned[["sex","age_max_live_or_dead_y","weight_date","taxon"]]
```

```python
df_cleaned[df_cleaned['weight_date'].notna()]
```

```python
# 2. Extract years from 'weight_date' and create a new column
df_cleaned['weight_year'] = pd.to_datetime(df_cleaned['weight_date'], format='%d/%m/%Y').dt.year
```

```python
df_cleaned["weight_year"]
```

```python
duplicates = df_cleaned[df_cleaned.duplicated(subset=['name', 'dob',"weight_year"], keep=False)]
print("Rows that are duplicated:")
print(duplicates)
```

```python
# 3. Keep only the last occurrence of duplicates and remove others
df_cleaned = df_cleaned.drop_duplicates(subset=['name','dob',"weight_year"], keep='last')

# 4. Verify the results
print("\nNumber of rows before removing duplicates:", len(df))
print("Number of rows after removing duplicates:", len(df_cleaned))
```
